```python
def y_axis_label(fig_args):
    y = fig_args['y']
    if 'pi' in y:
        return f'{pi} Diversity'
    elif 'diverg' in y:
        if 'rate' in y:
            return 'Divergence Rate'
        else:
            return 'Divergence'
    elif 'muta' in y:
        return 'Mutation Rate'
```
